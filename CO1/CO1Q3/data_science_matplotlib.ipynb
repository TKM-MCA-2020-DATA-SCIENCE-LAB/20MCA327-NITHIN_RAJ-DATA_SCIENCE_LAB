```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
x
y = np.random.randn(10)
y
```

```python
plt.plot(x,y,'r')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'r--')
plt.subplot(2,2,4)
plt.plot(x,y,'g*-')
```
