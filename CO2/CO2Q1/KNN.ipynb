```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/sonar_csv.csv')
```

```python
df
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
scalar.fit(df.drop('Class',axis=1))
```

```python
scaled_feature = scalar.transform(df.drop('Class',axis=1))
```

```python
a = pd.DataFrame(scaled_feature,columns=df.columns[:-1])
a
```

```python
from sklearn.model_selection import train_test_split
X = a
Y = df['Class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k = KNeighborsClassifier(n_neighbors=1)
```

```python
k.fit(x_train,y_train)
```

```python
pred = k.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from pandas.io.pytables import AppendableMultiSeriesTable
#Analyzing better k value through iterations
error_rate=[]

for i in range(1,40):
  k = KNeighborsClassifier(n_neighbors=i)
  k.fit(x_train,y_train)
  pred_i = k.predict(x_test)
  error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',marker='s',markerfacecolor='orange',markersize=10)
plt.title('Error rate vs K values')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
k = KNeighborsClassifier(n_neighbors=3)
k.fit(x_train,y_train)
pred = k.predict(x_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```
